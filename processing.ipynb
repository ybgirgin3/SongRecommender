```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
filename = 'LITTLE BIG â€“ FARADENZA (official music video).mp4'
```

```python
plt.figure(figsize = (15,4))
data, sample_rate = librosa.load(filename, sr=22050, mono=True, offset=0.0, duration=50, res_type='kaiser_best')
librosa.display.waveplot(data, sr=sample_rate, max_points=50000.0, x_axis=True, offset = 0.0, max_sr = 1000)
```
